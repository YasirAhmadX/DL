## Read Data

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dataset.csv');
df
```

```python
t = df['Text'].values
l = df['Label'].values
t
```

```python
l
```

## Feature Extraction

```python
t,l
```

### CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split

cv = CountVectorizer()
f = cv.fit_transform(t)

print(f)
```

### TfidfVectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vt = TfidfVectorizer()
ft = vt.fit_transform(t)

print(ft)
```

## MLP

```python
from sklearn.neural_network import MLPClassifier

ml = MLPClassifier()

ml.fit(f,l)
```

```python
s1 = ["The movie was not bad"]
st1 = cv.transform(s1)

ml.predict(st1)
```

```python
s2 = ["The movie was not good"]
st2 = cv.transform(s2)

ml.predict(st2)
```

```python
s3 = ["The movie is good"]
st3 = cv.transform(s3)

ml.predict(st3)
```

### SVM

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(f,l)
```

```python
print(s1)
print(svc.predict(st1))

print(s2)
print(svc.predict(st2))

print(s3)
print(svc.predict(st3))
```

### Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB

nb = MultinomialNB()
nb.fit(f, l)
```

```python
print(s1)
print(nb.predict(st1))

print(s2)
print(nb.predict(st2))

print(s3)
print(nb.predict(st3))
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()
rfc.fit(f,l)
```

```python
print(s1)
print(rfc.predict(st1))

print(s2)
print(rfc.predict(st2))

print(s3)
print(rfc.predict(st3))
```
